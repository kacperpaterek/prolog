# **ZADANIE 1**

```python
% Deklaracja discontiguos została użyta, aby  definiować klauzule predykatu w różnych częściach programu
:- discontiguous mezczyzna/1.
       :- discontiguous kobieta/1.
:- discontiguous ojciec/2.
:- discontiguous matka/2.
:- discontiguous malzenstwo/2.

%FAKTY rodzina 1 i relacja rodziców
ojciec(adam, ewa).
ojciec(adam, szymon).
matka(zofia, ewa).
matka(zofia, szymon).
malzenstwo(adam, zofia).

%fakty rodzina 2 i relacja rodziców
ojciec(jan, anna).
ojciec(jan, piotr).
matka(maria, anna).
matka(maria, piotr).
malzenstwo(jan, maria).

%Fakty dziadkowie
ojciec(denis, zofia).
ojciec(denis, maria).
matka(natalia, zofia).
matka(natalia, maria).
mezczyzna(denis).
kobieta(natalia).


%fakty płeć obie rodziny
mezczyzna(adam).
kobieta(ewa).
mezczyzna(szymon).
kobieta(zofia).
mezczyzna(jan).
kobieta(maria).
mezczyzna(piotr).
kobieta(anna).

%Reguły rodzic
rodzic(X, Y) :- ojciec(X, Y).
rodzic(X, Y) :- matka(X, Y).
%Przykladowe pytanie: rodzic(jan, anna).
%Przykladowa odpowiedz: true.

%Reguła brat
brat(X, Y) :-
    rodzic(Z, Y),
    rodzic(Z, X),
 X\= Y, %  jeśli jest bratem nie może być kobietą
    mezczyzna(X).
%przykladowe pytanie: brat(piotr, anna).
%przykladowa odpowiedz: true.

%Reguła siostra
siostra(X, Y) :-
    rodzic(Z, Y),
    rodzic(Z, X),
X\= Y, %jeśli jest siostrą to nie może być mężczyzną
    kobieta(X).
%Przykladowe pytanie: siostra(ewa, szymon).
%przykladowa odpowiedz: true.

% Reguła rodzeństwa potrzebna do reguły kuzyna
rodzenstwo(X, Y) :-
  brat(X, Y).
rodzenstwo(X, Y) :-
  siostra(X, Y).

% Reguła kuzyn
kuzyn(X, Y) :-
  rodzic(Z, X), %rodzic z 1 rodziny
  rodzic(V, Y), %rodzic z 2 rodziny
  rodzenstwo(Z, V), %pokrewienstwo miedzy rodzicami z obu rodzin
    mezczyzna(X),
    X \= Y. %kuzyn nie moze byc kobieta
%Przykladowe pytanie: kuzyn(X, piotr).
%przykladowe odpowiedz: X = szymon.

%Regula kuzynka
kuzynka(X,Y) :-
   rodzic(Z, X), %rodzic z 1 rodziny
  rodzic(V, Y), %rodzic z 2 rodziny
  rodzenstwo(Z, V), %pokrewienstwo miedzy rodzicami z obu rodzin
    kobieta(X),
    X \= Y. %kuzynka nie moze byc mezczyzna
%Przykladowe pytanie: kuzynka(X, anna).
%Przykladowa odpowiedz: X = ewa.

```

# **ZADANIE 2**

```python
%Fakty loty
lotnisko(warszawa).
lotnisko(krakow).
lotnisko(gdansk).
lotnisko(lublin).
lotnisko(london).

%fakty połączenia, cena
polaczenie(warszawa, krakow, 300).
polaczenie(krakow, gdansk, 200).
polaczenie(warszawa, gdansk, 500).
polaczenie(lublin, warszawa, 1000).
polaczenie(krakow, lublin, 550).
polaczenie(lublin, gdansk, 100).
polaczenie(london, warszawa, 50).
polaczenie(gdansk, london, 300).

%Reguła połączeń, dziala jezeli polaczenie bezposrednie istnieje
polaczenie(X, Y) :- polaczenie(X, Y, _).
%przykladowe pytanie: polaczenie(lublin, warszawa).
%przykladowe odpowiedz: true.

%Reguła kosztow podrozy
koszt_podrozy(X, Y, C) :- polaczenie(X, Y, C).
%przykladowe pytanie: koszt_podrozy(lublin, gdansk, C).
%przykladowa odpowiedz: C = 100.


%Regula dotyczaca znalezienia najkrotszego i najtanszego polaczenia bezposredniego
najkrotsza_droga(X, Y) :-
    polaczenie(X, Y, C), % polaczenie bezposrednie
    format('Trasa bezposrednia, Cena: ~w~n', [C]). %to, co ma sie wyswietlic
%przykladowe pytanie: najkrotsza_droga(london, warszawa).
%przykladowa odpowiedz: Trasa bezposrednia, Cena: 50.

%Regula dotyczaca najkrotszego i najtanszego polaczenia z przesiadka
najkrotsza_droga(X, Y) :-
    polaczenie(X, Z, C1), %polaczenie z lotniska poczatkowego do lotniska posredniego
    polaczenie(Z, Y, C2), %polaczenie z lotniska posredniego do lotniska docelowego
    C is C1 + C2, %sumowanie calkowitych  kosztów polaczenia
    format('Trasa przez ~w, Koszt: ~w~n', [Z, C]). % to, co ma sie wyswietlic
%przykladowe pytanie: najkrotsza_droga(krakow, london).
%przykladowa odpowiedz: Trasa przez gdansk, koszt 500.



```

# **ZADANIE 4**

```python
%Fakty miasta
miasto(lublin).
miasto(poznan).
miasto(katowice).
miasto(parczew).
miasto(radom).

%fakty polaczenia drogowe i czas przejazdu
droga(lublin, poznan, 5).
droga(poznan, katowice, 4).
droga(katowice, parczew, 5).
droga(parczew, radom, 2).
droga(lublin, katowice, 4).
droga(lublin, parczew, 10).
droga(radom, lublin, 1).
droga(poznan, parczew, 5).
droga(poznan, radom, 4).
droga(katowice, radom, 3).
droga(radom, katowice, 3).

%regula dla sprawdzenia czy istnieja bezposrednie polaczenia miedzy miastami
droga(X,Y) :-
    droga(X, Y, _). %jesli polaczenie istnieje "true", jesli nie "false"
 %przykladowe pytanie: droga(poznan, radom).
 %przykladowa odpowiedz: true


%regula dla okreslenia czasu przejazdu miedzy miastami jezeli istnieje bezposrednie polaczenie
czas_przejazdu(X, Y, T) :-
    droga(X, Y, T).
%przykladowe pytanie: czas_przejazdu(lublin, katowice, T).
%przykladowa odpowiedz: T=4


%jezeli istnieje tylko polaczenie posrednie miedzy miastami
czas_przejazdu(X, Y, T) :-
    droga(X, Z, T1), %droga z punktu startowego do punktu posredniego
    czas_przejazdu(Z, Y, T2), %droga z punktu posredniego do koncowego
    T is T1 + T2. %calkowity czas przejazdu
%przykladowe pytanie: czas_przejazdu(katowice, lublin, T).
%przykladowa odpowiedz: T=8

%Regula na znalezienie najkrotszego czasu przejazdu miedzy miastami
najkrotsza_trasa(X,Y) :-
      droga(X, Z, T1), %trasa z miasta poczatkowego do miasta posredniego
      droga(Z, Y, T2), %trasa z miasta posredniego do miasta docelowego
    T is T1 + T2, %sumowanie calkowitego czasu przejazdu
    format('Trasa przez ~w, czas ~w~n', [Z, T]). % to, co ma sie wyswietlic
%przykladowe pytanie: najkrotsza_trasa(katowice, lublin).
%przykladowa odpowiedz: Trasa przez radom, Czas 4





```

```python
:- dynamic ocena/3. %dynamiczne dodawanie i usuwanie ocen
%fakty ksiazka, autor, gatunek
ksiazka('Wiedzmin', sapkowski, fantasy).
ksiazka('Hobbit', tolkien, fantasy).
ksiazka('Makbet', szekspir, dramat).
ksiazka('Balladyna', slowacki, dramat).
ksiazka('Opowiesci z Narnii', lewis, fantasy).
ksiazka('Tango', mrozek, dramat).

%fakty oceny
ocena(uzytkownik1, 'Wiedzmin', 5).
ocena(uzytkownik1, 'Makbet', 1).
ocena(uzytkownik1, 'Opowiesci z Narnii', 4).
ocena(uzytkownik2, 'Hobbit', 2).
ocena(uzytkownik2, 'Tango', 5).

%fakty preferencje
preferencje(uzytkownik1, fantasy).
preferencje(uzytkownik2, dramat).

%dodawanie oceny
dodaj_ocene(Użytkownik, Tytuł, Ocena) :- %Usuwa istniejącą ocenę a następnie dodaje nową
    retractall(ocena(Użytkownik, Tytuł, _)), %usuwa ocene
    assertz(ocena(Użytkownik, Tytuł, Ocena)), %dodaje nowa ocene
    format('Dodano ocene~n'). %tekst do wyswietlenia
%przykladowe pytanie: dodaj_ocene(uzytkownik2, 'Hobbit', 5).
%przykladowa odpowiedz: Dodano ocene.

%usuwanie oceny
usun_ocene(Użytkownik, Tytul) :- %Usuwa ocenę dla danej książki
    retract(ocena(Użytkownik, Tytul, _)), %usuwa ocene
    format('Usunieto ocene~n'). %tekst do wyswietlenia
%przykladowe pytanie: usun_ocene(uzytkownik2, 'Hobbit').
%przykladowa odpowiedz: Usunieto ocene.

%Regula rekomendacji: identyfikowanie ksiazek po gatunku i rekomendowanie ich do osob o konkretnych preferencjach
polecane_ksiazki(X, Gatunek) :-
    findall(
        Tytul,
        ( ksiazka(Tytul, _, Gatunek),      % Zbiera wszystkie możliwe tytuly ksiazek
          preferencje(X, Gatunek),        % Gatunek pokrywa się z preferencjami użytkownika,
          ocena(X, Tytul, Ocena),         % Tytuly z danego gatunku wybrane na podstawie ocen użytkownika
          Ocena >= 4                      % Ocena ma mieć wartość większą lub równą 4
        ),
        PolecaneKsiazki                   % W liście PolecaneKsiazki
    ),
    format('Polecane książki: ~w~n', [PolecaneKsiazki]). %tekst do wyswietlenia
% Przykładowe pytanie: polecane_ksiazki(uzytkownik1, fantasy).
% przykladowa odpowiedz: Polecane książki: [Wiedzmin,Opowiesci z Narnii]
```
